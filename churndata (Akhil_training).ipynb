# Discover outliers with mathematical function (Z-Score)


```python
from scipy import stats
import numpy as np
```

```python
z = np.abs(stats.zscore(A[['msisdn','accountLength','areaCode','numberVmailMessages','totalDayMinutes','totalDayCalls',
                           'totalDayCharge','totalEveMinutes','totalEveCalls','totalEveCharge','totalNightMinutes',
                           'totalNightCalls','totalNightCharge','totalIntlMinutes','totalIntlCalls','totalIntlCharge',
                           'numberCustomerServiceCalls']]))
print(z)
```

```python
#Looking the code and the output above, it is difficult to say which data point is an outlier.
#Letâ€™s try and define a threshold to identify an outlier.

threshold = 3
print(np.where(z > 3))
```

```python
#Donâ€™t be confused by the results. The first array contains the list of row numbers and second array respective column numbers,
#which mean z[22][14] have a Z-score higher than 3 and so on ....

print(z[22][14])
print(z[32][7])
print(z[32][9])
```

# Removal of outliers through z score 

```python
A_o = A[(z < 3).all(axis=1)]    
```

# Decision Tree 

```python
X=A_o[['totalDayMinutes','totalDayCalls','totalEveMinutes','totalNightMinutes',
       'totalIntlMinutes',
       'numberCustomerServiceCalls','totalIntlCalls','voiceMailPlan_Dummy']]

Y=A_o[["churner_Dummy"]]

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2,random_state=20)

from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier(criterion="entropy",min_samples_leaf=7)
model_dtc=dtc.fit(xtrain,ytrain)
pred_dtc=model_dtc.predict(xtest)
from sklearn.metrics import accuracy_score
print(accuracy_score(ytest['churner_Dummy'],pred_dtc))
print(confusion_matrix(ytest['churner_Dummy'],pred_dtc))
```
